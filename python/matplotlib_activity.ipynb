Draw four plots in one figure

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
names = ['Tony', 'Joe', 'Sally', "Andrea"]
num_marbles = [100, 500, 150, 10]

fig, axs = plt.subplots(2, 2,figsize=(20, 16))
axs[0, 0].plot(names, num_marbles, 'tab:orange')
axs[0, 0].set_title('Line')
axs[0, 1].bar(names, num_marbles)
axs[0, 1].set_title('Bar')
axs[1, 0].scatter(names, num_marbles, s=[num_marbles[num]*5 for num in range(len(num_marbles))])
axs[1, 0].set_title('Scatter')
axs[1, 1].pie(num_marbles, labels=names)
axs[1, 1].set_title('Pie')


# Need to set the y limits on the scatter plot so the top
# marker does not get sut off.
axs[1, 0].set_ylim([0,550])

for ax in axs.flat:
    ax.set(xlabel='Name', ylabel='# Marbles')
    
#for ax in axs.flat:
#    ax.label_outer()
    
axs[1, 1].set(xlabel='# Marbles', ylabel='')

fig.tight_layout()

SMALL_SIZE = 28
plt.rc('font', size=SMALL_SIZE)
plt.rc('axes', titlesize=SMALL_SIZE)

plt.style.use('fivethirtyeight')

axs[0, 0].grid(b=False)
axs[0, 1].grid(b=False)
axs[1, 0].grid(b=False)

plt.subplots_adjust(hspace=.2)

plt.savefig("Images/four_plots_exercise.png")

plt.plot()
```
